```python
data_with_missing ={
    "cars": ["BMW", "Volvo", None, "Bugatti", "Ferarri", None, "Porsche"],
    "counts": [None, 7, 5, 4, 3, None, 9]
}

df_missing = pd.DataFrame(data_with_missing)
print(df_missing)
```

```python
df_missing.isnull()
```

```python
df_missing.isnull().sum()
```

```python
df_missing.dropna()
```

```python
df_missing.dropna(how="all")
```

```python
df_missing.dropna(axis=1)
```

```python
df_missing.dropna(subset=["counts"]) # faqat counts columns ni tekshiradi!
```

```python
df_missing.fillna(0) # None qatorlarga data joylaydi
```

```python
df_missing.fillna({"cars": "unknow", "counts": 0}) # None qatorlarga data joylaydi
```
