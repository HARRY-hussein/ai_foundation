```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print("Dataset shape (rows, columns):", df.shape)
print("Column name:", df.columns.tolist())

print("First 5 rows of the dataset:")
print(df.head())

print("Summary statistics:")
print(df.describe())

print("Data types and null info:")
print(df.info())
```

```python
print("Average scores by subject:")
df[["math score", "reading score", "writing score"]].mean()
```

```python
print("Maximum scores by subject:")
df[["math score", "reading score", "writing score"]].max()
```

```python
print("Minimum scores by subject:")
df[["math score", "reading score", "writing score"]].min()
```

```python
# Correlation
print("Correlation:")
df[["math score", "reading score", "writing score"]].corr()
```

```python
# New columns creation
df["total score"] = df[["math score", "reading score", "writing score"]].sum(axis=1)
df["average score"] = df["total score"] / 3
df["passed"] = np.where(df["average score"] >= 50, "Yes", "No")

print("New columns preview (total, average, passed):")
print(df[["total score", "average score", "passed"]].head())
```

```python
test_prep_group = df.groupby('test preparation course')[["math score", "reading score", "writing score"]].mean()
print("student score by test preperation course:")
print(test_prep_group)

# Gender-based average
gender_avg = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
print("Average score by gender:")
print(gender_avg)

# Who passed vs failed
pass_rate = df["passed"].value_counts(normalize=True) * 100
print("Pass rate:\n", pass_rate)
```

```python
# 'math score', 'reading score' va 'writing score' ustunlarini tanlab,
# ularni Numpy massiviga (array) aylantirib oldik
scores = df[["math score", "reading score", "writing score"]].values

# # Mean (o'rtacha qiymat) va Std Deviation (standart chetlanish)
# Har bir ustun (axis=0) bo'yicha o'rtacha qiymatni hisoblaymiz
print("Numpy Mean:", np.mean(scores, axis=0))
# Har bir ustun bo'yicha standart chetlanishni hisoblaymiz
print("Numpy Std:", np.std(scores, axis=0))

# # Z-score (normallashtirish)
# Z-score formulasi: (qiymat - o'rtacha) / standart chetlanish
# Bu ma'lumotlarni o'lchov birliklaridan xalos etib, taqqoslashni osonlashtiradi
z_score = (scores - np.mean(scores, axis=0)) / np.std(scores, axis=0)

print("Z-scores (normalized numbers)")
# Normallashtirilgan ma'lumotlarning dastlabki 5 tasini chiqaramiz
print(z_score[:5])

# Izohlar:
# Average 0  -> Normallashtirilgandan so'ng, o'rtacha qiymat 0 ga teng bo'ladi
# Negative - -> O'rtacha qiymatdan past bo'lgan natijalar
# Positive + -> O'rtacha qiymatdan yuqori bo'lgan natijalar
```

```python
# O'rtacha bali 90 va undan yuqori bo'lgan talabalar
# 'average score' ustunidan 90 dan katta yoki teng bo'lgan qatorlarni filtrlash
top_scorers = df[df["average score"] >= 90]
# top_scorers massividagi elementlar sonini (len) hisoblab chiqaramiz
print("Top scored student above 90:", len(top_scorers))

# Imtihondan o'ta olmaganlar sonini aniqlash
# 'passed' ustunida "No" (o'ta olmagan) bo'lgan qatorlarni filtrlash
failed = df[df["passed"] == "No"]
# failed massividagi elementlar sonini (len) hisoblab chiqaramiz
print("Failed from exam:", len(failed))
```

```python
# 'passed' ustuni bo'yicha guruhlab, fanlardan average ballarni hisoblash
comparison = df.groupby("passed")[["math score", "reading score", "writing score"]].mean()

print("Average scores of both passed and failed students by subject:")
print(comparison)
```

```python
# "passed" ustunidagi barcha qiymatlar
print("All ranking:")
print(df["passed"].value_counts(dropna=False))
```

```python
# Faqat yiqilganlar guruhi
failed_students = df[df["passed"] == "No"]

# Average scores of failed students by subject
print("Average scores of failed students by subject:")
print(failed_students[["math score", "reading score", "writing score"]].mean())
```

```python
# 1. Dividing students into three ranking
# above 80 - "Excellent", 50-80 - "Average", below 50 - "Weak"
def categorize_student(score):
    if score >= 80:
        return 'Excellent'
    elif score >= 50:
        return 'Average'
    else:
        return 'Weak'

# apply() funksiyasi orqali har bir qatorga funksiya qo'llanadi
df['performance_category'] = df['average score'].apply(categorize_student)

# 2. Har bir guruhda nechta talaba bor
category_counts = df['performance_category'].value_counts()
print("Distribution of students by performance level:")
print(category_counts)

# 3. Har bir guruhning matematikadan o'rtacha bali (Taqqoslash uchun)
performance_group_mean = df.groupby('performance_category')[['math score']].mean()
print("\nAverage math score for each performance group:")
print(performance_group_mean)
```

```python
# "Lunch" va "Test preparation course" holatidagi o‘rtacha ballar
pivot_analysis = df.pivot_table(
    index='lunch',
    columns='test preparation course',
    values='average score',
    aggfunc='mean'
)

print("Impact of lunch and test preparation course on average scores:")
print(pivot_analysis)

# 2. Xulosa chiqarish uchun farq
# Kursga borgan va bormaganlar o'rtasidagi farq
diff = pivot_analysis['completed'] - pivot_analysis['none']
print("\nAverage score difference between students who completed the course and those who did not (in points):")
print(diff)
```
